# Download symbols from SP500

Many ways to acieve this.

I am using a simple web scrape from this wiki pedia page https://en.wikipedia.org/wiki/List_of_S%26P_500_companies

the `pandas.read_html` returns a list of DataFrames - one for each HTML table in the source web site.

The data needed to this is the first table/DataFrame

```python
import pandas as pd

df_sp500 = pd.read_html(io='https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
```

```python
# Have a look at what was downloaded
pd.concat(objs=[df_sp500.head(), df_sp500.tail()])
```

Download Open, High, Low, Close for symbls in question.

In this example, use `yfinance` to download OHLC from yyahoo finance and store in the dfirectory `./data`

```python
import yfinance as yf
import os

from datetime import datetime
from dateutil.relativedelta import relativedelta

today = datetime.now()
last_year = today - relativedelta(years=1)

to_ = today.strftime('%Y-%m-%d')
from_ = last_year.strftime('%Y-%m-%d')

dir_ = '../Data/Daily/'
if not os.path.isdir(dir_):
    os.mkdir(dir_)

for ticker in df_sp500['Symbol']:
    df = yf.download(tickers=ticker, start=from_, end=to_)
    file_ = f'{dir_}{ticker}.csv'
    df.to_csv(file_)
    
```

Save the Sp500 list of companies to the `./Data` directory

```python
cols = ['Symbol','Security']
df_sp500[cols].to_csv(path_or_buf='../Data/sp500.csv', index = False)
```
